# Assignment 2: Stock Market Analysis Agent

This notebook demonstrates the LangGraph agent built for Assignment 2.

```python
from IPython.display import Markdown, display
import sys
import os
sys.path.append(os.path.join(os.getcwd(), 'stock_agent'))

from stock_agent.graph import stock_agent_app

def run_agent(ticker: str):
    print(f"\n{'='*50}")
    print(f"Running Analysis for: {ticker}")
    print(f"{'='*50}\n")
    
    # Initialize state
    initial_state = {
        "ticker": ticker,
        "historical_data": None,
        "current_price": None,
        "sma_10": None,
        "sma_20": None,
        "rsi_14": None,
        "recommendation": None,
        "report": None,
        "error": None
    }
    
    # Run the graph
    result = stock_agent_app.invoke(initial_state)
    
    # Display the formatted report
    display(Markdown(result['report']))
```

### Test Case: Valid Ticker

```python
run_agent("AAPL")
```
